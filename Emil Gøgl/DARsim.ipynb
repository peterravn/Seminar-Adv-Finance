```python
# Parameters

T = 1000
pi = 0
alpha = 0
omega = 1
```

```python
# Simulate DAR process (random coeffecient form)
def DAR_sim(pi, alpha, omega, x_0, T):
    
    # Generate random variables
    phi = np.random.normal(loc = 1 + pi, scale = np.sqrt(alpha), size = T)
    eta = np.random.normal(loc = 0 + pi, scale = np.sqrt(omega), size = T)
    
    # Define empty varaible(s)
    x = np.zeros(T)
    
    # Initialise process
    x[0] = x_0
    
    for t in range(1,T):
        x[t] = phi[t] * x[t-1] + eta[t]
    return x

T = 500

series1 = DAR_sim(0, 0, 1, 0, T)
series2 = DAR_sim(0, 1, 1, 0, T)
```

```python
plt.subplot(1, 2, 1)
plt.plot(series1, label = "alpha_0 = 0")
plt.legend()
plt.title("Random walk")

plt.subplot(1, 2, 2)
plt.plot(series2, label = "alpha_0 = 1")
plt.legend()
plt.title("Volatility induced stationarity")

save_path = "/Users/emilbisiach/Desktop/Polit KU/Kandidat/1. semester/Financial Econometrics/Assingments/Assignment 1/"
#plt.savefig(save_path + "DAR simulation.png")
```
